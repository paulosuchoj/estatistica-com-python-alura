### Para avaliarmos o comportamento da variável RENDA vamos construir uma tabela de frequências considerando as seguintes classes em salários mínimos (SM)
#### <font color='blue'>Descreva os pontos mais relevantes que você observa na tabela e no gráfico.</font>

Classes de renda:

<b>A</b> ► Acima de 25 SM

<b>B</b> ► De 15 a 25 SM

<b>C</b> ► De 5 a 15 SM

<b>D</b> ► De 2 a 5 SM

<b>E</b> ► Até 2 SM

Para construir as classes de renda considere que o salário mínimo na época da pesquisa era de <b>R$ 788,00</b>.

#### Siga os passos abaixo:

### 1º Definir os intevalos das classes em reais (R$)

```python
classes = [df.Renda.min(), 2 * 788, 5 * 788, 15 * 788, 25 * 788, df.Renda.max()]
```

### 2º Definir os labels das classes

```python
labels = ['E', 'D', 'C', 'B', 'A']
```

### 3º Construir a coluna de frequências

```python
frequencias = pd.value_counts(
    pd.cut(
        x = df.Renda,
        bins = classes,
        labels = labels,
        include_lowest = True
    )
)

frequencias
```

### 4º Construir a coluna de percentuais

```python
percentuais = pd.value_counts(
    pd.cut(
        x = df.Renda,
        bins = classes,
        labels = labels,
        include_lowest = True
    ), normalize = True
) * 100

percentuais
```

### 5º Juntar as colunas de frequência e percentuais e ordenar as linhas de acordo com os labels das classes

```python
dados_frequencias_percentuais = pd.DataFrame({'Frequências' : frequencias, 'Percentuais' : percentuais}).sort_index(ascending=False)
dados_frequencias_percentuais
```

### Construa um gráfico de barras para visualizar as informações da tabela de frequências acima

```python
ax = sns.barplot(data=dados_frequencias_percentuais, x = dados_frequencias_percentuais.index, y=dados_frequencias_percentuais.Frequências)
```

```python
renda_classe_a = df.query("Renda > 19700")
renda_demais_classes = df.query("Renda <= 19700")

num_todas_classes = df.Renda.count()
num_classe_a = df.query("Renda > 19700").Renda.count()
num_classe_demais_classes = df.query("Renda <= 19700").Renda.count()

soma_de_rendas_total = df.Renda.sum()
soma_de_rendas_altas = renda_classe_a.Renda.sum()
soma_de_rendas_baixas = renda_demais_classes.Renda.sum()

moeda_renda_total = lc.currency(soma_de_rendas_total, grouping=True).replace('$', 'R$').replace(',', '.').replace('.00', ',00')
moeda_rendas_altas = lc.currency(soma_de_rendas_altas, grouping=True).replace('$', 'R$').replace(',', '.').replace('.00', ',00')
moeda_rendas_baixas = lc.currency(soma_de_rendas_baixas, grouping=True).replace('$', 'R$').replace(',', '.').replace('.00', ',00')

print(f"A renda total mensal da população entrevistada na PNAD é de {moeda_renda_total}")
print(f"Um total de {num_classe_a} pessoas concentra {moeda_rendas_altas} em renda mensal")
print(f"O restante, {num_classe_demais_classes} concentra {moeda_rendas_baixas} em renda mensal")
```

```python
porcentagem_classe_a = "{:.2%}".format(num_classe_a / num_todas_classes)
porcentagem_renda_classe_a = "{:.2%}".format(soma_de_rendas_altas / soma_de_rendas_total)

print(f"A classe A representa {porcentagem_classe_a} da amostra total de {num_todas_classes} entrevistadas. Ela concentra {porcentagem_renda_classe_a} da renda mensal total de {moeda_renda_total}.")
```

### Obtenha a média, mediana e valor máximo da variável RENDA segundo SEXO e COR

```python
renda_descritiva_por_sexo_e_cor = pd.crosstab(df.Cor, df.Sexo, values = df.Renda, aggfunc = {'mean', 'median', 'max'})
renda_descritiva_por_sexo_e_cor.rename(index = cor, inplace = True)
renda_descritiva_por_sexo_e_cor.rename(columns = sexo, inplace = True)
renda_descritiva_por_sexo_e_cor
```

### Obtenha as medidas de dispersão da variável RENDA segundo SEXO e COR
#### <font color='blue'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mad', 'var', 'std'}</i></font>

```python
renda_dispersao_por_sexo_e_cor = pd.crosstab(df.Cor, df.Sexo, values = df.Renda, aggfunc = {'mad', 'var', 'std'}).round(2)
renda_dispersao_por_sexo_e_cor.rename(index = cor, inplace = True)
renda_dispersao_por_sexo_e_cor.rename(columns = sexo, inplace = True)
renda_dispersao_por_sexo_e_cor
```

O desvio médio absoluto dos homens e mulheres pretos é o mais próximo um do outro, o que significa que eles são uma população que em geral ganha muito próximo da média da classificação.

O método <i>agg()</i> pode receber um dicionário especificando qual coluna do DataFrame deve ser utilizada e qual lista de funções estatísticas queremos obter, por exemplo: <i>dados.groupby(['UF']).agg({'Renda': ['mean', 'median', 'max', 'std']})</i></font>
